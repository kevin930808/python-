```python
import os
import cv2
path='./'
imgs=os.listdir(path)
n=1
for img in imgs:
    print(img)
    i = cv2.imread('%s'%img)
    cv2.imshow('os',i)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    newpath=str(n)+'.jpg'
    print(newpath)
    cv2.imwrite('new%s'%newpath,i)
    n=n+1
    new = cv2.flip(i,1)
    newpath=str(n)+'.jpg'
    cv2.imwrite('new%s'%newpath,new)
    n=n+1
```
