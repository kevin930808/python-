```python
import random 
import numpy as np
from math import sqrt
random.seed(10)
M=[]
for i in range(230):
    x=random.randint(1,6)
    M.append(x)
print(M)
##
def listSum(list):
    return sum(list)
def listMean(list):
    Mean = sum(list)/len(list)
    return Mean
def listDev(list,up=0,std=0):
    x_bar=listMean(list)
    for i in range(len(list)):
        up = up + (list[i]-x_bar)**2
    std = sqrt((up/(len(list)-1)))
    return std
print(listSum(M),listMean(M),listDev(M))
print(np.sum(M),np.mean(M),np.std(M))
def change(list,row,colum):
    N=[]
    addtime=0
    i=0
    j=0
    N.append([])
    while j<230:
        if addtime >= colum:
            N.append([])
            addtime =0
            i = i+1
        N[i].append(list[j])
        addtime+=1
        j=j+1
    if len(N)==row:
        return N

print(change(M,23,10))
N= change(M,23,10)

def rowsum(list):
    ans=[]
    for i in range(len(list)):
        ans.append(sum(list[i]))
    return ans
print(rowsum(N))
def columsum(list):
    ans=[]
    M10_23=change(M,10,23)
    for i in range(len(M10_23)):
        ans.append(sum(M10_23[i]))
    return ans
print(columsum(M))   
M=np.array(M).reshape(10,23)
print(np.sum(M,axis=1))
    
```

```python
import random
random.seed(10)

M = []
for i in range(0,230):
    n = random.randint(1,5)
    M.append(n)
M[99:104]#4、2、1、5、1

nc = []
for i in range(1,6):
    nc.append(M.count(i))
    print("Count of", i, "=", nc[i-1])
#39、 47、 42、 61、 41

def listSum(x):
    sum = 0
    for value in x:
        sum = sum + value
    return sum
listSum(M)#708
        
def listMean(x):
    sum = 0
    for value in x:
        sum = sum + value
    return round(sum / len(x),2)
listMean(M)#3.08

import math
def listDev(x):
    average = listMean(x)
    squareSum = 0
    for value in x:
        squareSum =squareSum + (value - average) ** 2
    return round(math.sqrt(squareSum / (len(x) - 1)),2)
listDev(M)#1.36

N = []
start = 0
end = 10
for row in range(0,23):
    N.append(M[start:end])
    start = start + 10
    end = start + 10# codes for generate N
len(N)   
     
def rowSum(m):
    row = []
    for i in range(0,23):
        row.append(listSum(m[i]))
    return row
rowSum(N)
listSum(rowSum(N))#708

def sumColumn(m):
    col = []
    colSum = []
    for i in range(0,10):
        new = []
        for j in range(0,23):
            new.append(m[j][i])
        col.append(new)# transpose matrix m to col
        colSum.append((listSum(col[i])))
    return colSum
sumColumn(N)
listSum(sumColumn(N))#708
```
